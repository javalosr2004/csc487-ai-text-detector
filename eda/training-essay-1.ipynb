```python
import pandas as pd

# Path is relative to this notebook's location
file_path = "../data/Training_Essay_Data.csv"

# Read the CSV into a DataFrame
df = pd.read_csv(file_path)


```

```python
ai_df = df[df['generated'] == 1]
human_df = df[df['generated'] == 0]

# Take a random sample of 10 rows with a fixed seed for reproducibility
ai_sample_df = ai_df.sample(n=10, random_state=42)
human_sample_df = human_df.sample(n=10, random_state=42)
```

```python
# Calculate text lengths (word count) for both AI and human samples
ai_sample_df['text_length'] = ai_sample_df['text'].str.split().str.len()
human_sample_df['text_length'] = human_sample_df['text'].str.split().str.len()

# Print summary statistics for AI text lengths
print("AI Text Length Summary (word count):")
print(f"Min: {ai_sample_df['text_length'].min()}")
print(f"Max: {ai_sample_df['text_length'].max()}")
print(f"Mean: {ai_sample_df['text_length'].mean():.2f}")
print()

# Print summary statistics for Human text lengths
print("Human Text Length Summary (word count):")
print(f"Min: {human_sample_df['text_length'].min()}")
print(f"Max: {human_sample_df['text_length'].max()}")
print(f"Mean: {human_sample_df['text_length'].mean():.2f}")
print()
```
